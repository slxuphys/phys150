# How does Python Handle Variables? Immutability vs Mutability

Everything in Python, including variables, is an object, associated with a unique id. Id is an integer that is unique for the given object and remains constant during its lifetime.

<details>
<summary> </summary>
<p>

 For scaler types, int, float, complex, string, none, the value associated to an id cannot be changed.
    
 Immutability!

````python
a='Hello!'
b=a
print('id a',id(a),'id b',id(b))

# a and b have the same id
#two different names of the same object
````

````python
a='physics'# a is no long pointed to 'hello' but assgined a new variable
print('a:',a,id(a))
print('b:',b,id(b))
````

|name of the variables| id of the variables | value of the variables|
|:----:|:---:|:---:|
 |~a~,b| 324...560| 'Hello'|
 |a|324...808|'Physics'|

For scaler variable, there is nothing one can do to change the value without changing the id number. This property is called immutable.

## Scaler Types are immutable. List is mutable.
List can be updated __in place__.

List is similar to a folder, and its elements are similar to files. One can modify the files without creating an entirely new folder

````python
a=[1,'323',True] #a list
print('a',a,id(a))
b=a #a, b point to the same object and have same id
c=[1,'323',True]
#c has the same elements, but is a different object
print('b',b,id(b))
print('c', c, id(c))
````

|name |id| value|
|:---:|:--:|:----:|
| a,b| 290...1968|[1 , '323',True]|
|c|290...2160|[1, '323',True]|

````python
a[0]=3 #updates a
print(a)
print(b) #it also updates b, because a and b point to the SAME object
print(c) #c remains unchanged because it points to a different object
````

- Attribute of List
    - __copy__ (makes a same list but different id)
    
    - mutable method (update the list in place)
    
        -  append, extend, insert, pop, remove, clear 
    
        -  reverse, sort
    
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The list is updated __in place__, and the id does not change.

````python
a=[1,1,2,3,'apple']
b1=a.copy()
b2=a
#b2 is the same object as a, b1 is a different object
print(a,id(a))
print(b1,id(b1))
print(b2,id(b2))
````

````python
a=[True,False,3,4]
print(a,id(a))
a.append('orange') #add a new element in a
print(a,id(a))
a.extend([1,2]) #extend a with another list, IN PLACE
print(a,id(a))
list_num=[10,3.0,4,2]
list_num.reverse() #reverse order
print(list_num)
````

## Set operations 

between two sets, a and b

- compare two sets

|equal|strict subset|subset|strict superset|superset|
|---|---|---|---|--|
| == | < | $\leq$| >| $\geq$|


- produce a new set


|union|intersection|difference|symmetric difference|
|--|---|--|--|
|\|| &|  -| ^ |

a and b are not changed



````python
set_1={'Einstein', 'Newton', 'Dirac'}
set_2={'Einstein', 'Heisenberg', 'Schrodinger'}
set_3={'Einstein'}
print(set_1)
print(set_2)
print(set_1==set_2)
print(set_3<set_1)
print(set_1>set_3)
````

````python
set_1={'Einstein', 'Newton', 'Dirac'}
set_2={'Einstein', 'Heisenberg', 'Schrodinger'}
#intersection: element in a and in b
set_intrstn = set_1 & set_2
print('set_intrstn:',set_intrstn)
#union: elements in a or in b
set_union = set_1 | set_2
print('set_union:',set_union)
#difference: elements in a but not in b
set_diff = set_1 - set_2
print('set_diff:', set_diff)
#symm. difference: union - intersection, not in a or not in b
set_symdiff = set_1 ^ set_2
print('set_symdiff:', set_symdiff)
````

# Numpy Array

Numpy is a python package that includes types and a lot of functions for manipulating vectors/matrix and vector calculations, e.g. linear algebra. 

It is one of the most popular python package. 

It defines a new type: array. (__different from a list__)

To use a package, we need first to install it and import it to the notebook. 

Because numpy is very popular, anaconda automatically installs it for you. 

To import numpy 

```python
    import numpy
```

to define an numpy array

```python
    a_array=numpy.array([1,3,4,4]) # the input is a list or a tuple, and the numpy.array function converts it to an array
```

Array behaves differently under + and \*

More details on packages later!

````python
#from package
import numpy # numpy is a common library
v1=[2,3]
v2=[1,4] # two lists
v3=v1+v2 # put the two list together
print(v3)
v1_array = numpy.array(v1) #an array
print(v1_array,type(v1_array))
v2_array = numpy.array(v2) # another array
print(v2_array,type(v2_array))
v3=v1_array+v2_array #sum the two array
print(v3)
v4=v1_array*5 #mutiply each elements in the array by 5
print(v4)
````
