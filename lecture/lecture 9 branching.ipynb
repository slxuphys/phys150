Given a number, print "it is positive" if it is positive, print "it is negative" if it is negative and print "it is zero" otherwise

```python
a=float(input('a number: '))
if a>0 :
    print('a is positive')
elif a<0 :
    print('a is negative')
else :
    print('a is zero')
print(a) #outside if block
```

1. Input a password. Print "invalid password" if it does not contain number or does not contain capital letters

```python
num="1234567890"
alpha="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
numset=set(num)
alphaset=set(alpha)

a=input("input a password: ")
password=set(a)

if password.isdisjoint(numset) or password.isdisjoint(alphaset):
    print("invalid password")
else:
    print("valid password")
```
