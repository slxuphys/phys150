## String Type

string has many attributes, which can be accessed using variable_name.

#### useful string methods 
- a.find()  _returns the index of a substring_
- a.replace() _replaces a substring with a new one_
- a.count() _counts the number of substring that appears_
- a.format(s) _insert the variables to the string  _

````python
a='abcdefg'
print(a.find('cde')) #c is the third character, the index of which is 2
print(a.replace('c','C')) #replace a substring with a new substring
a='aaaccd' 
print(a.count('a')) # counts the number of substrings in a string
a='the answer is {}'
print(a.format(42)) #insert the number 42 to the place of the brace
# insert multiple numbers
a='the answer is {number1}, the second answer is {number2}' 
print(a.format(number1=3,number2=4))
````


## String Type

#### string index & slicing
- single character: s[indx]

- substring 
  - s[start:end:step]
  
start: starting index for the substring

end: end index for the substring __+1__

step can be positve or negative

````python
s='0123456789' #the length of the string is 10
print(s[9]) # single number to access a single character;
#the index starts at 0
````

````python
(3+1j).real
````

````python
# access a substring
# two colons;three numbers
s='0123456789'
print(s[1:5:1]) #indx from 1 to 4 (not 5) with step 1
print(s[1:5:2]) #every other number from indx 1 to 4
print(s[1:5]) #if the step is one, the last column and number can be omitted
print(s[:5]) #if starting from the begining, the first number can be omitted
print(s[1:]) #if going all the way to end, the second number can be omitted


# step can be -negative
print(s[5:0:-1]) #indx frm 5 to 1(!), not 0
print(s[5::-1]) #indx from 5 to 0; the second number is omitted
print(s[::-1]) #the whole string to reverse order

# negative indx should be intepreted as length + negative indx
# useful if we do not know the lengh
print(s[10-5:10:1]) #last 5 characters; most formal method
print(s[-5:]) #the first '10' is omitted becasue -5 is a negative indx
# the second '10' is omitted because it goes only the way to the end
# the third number is omitted because the step is 1.
````

###  Caculating the gravity force between two objects

\begin{equation}
F = G \frac{m_1 m_2}{r^2}
\end{equation}

$G=6.67\times 10^{-11} \frac{N kg^2}{m^2}$

````python
G=6.67e-11 #newton's constant in SI unit
m1=input('the mass of the first object in kg:')
m2=input('the mass of the second object in kg :')
r =input('the distsance between them in m:')
m1=float(m1) # convert to a number
m2=float(m2) # convert to a number
r =float(r)  # convert to a number
f = G * m1 * m2/(r**2)
s = 'OUTPUT:The force between the two objects is {} N';
print(s.format(f))
````
