### Netwon's law of gravitation
[Wikipedia page](https://en.wikipedia.org/wiki/Newton%27s_law_of_universal_gravitation)

In today's language, the law states that every point mass attracts energy other pint mass by a force acting along the line intersecting the two points. The force is proportional to the product of the two masses, and inversely proportional to the square of the distance between them. 

The equation for universal gravitation thus takes the form:
\begin{equation}
F=G \frac{m_1m_2}{r^2}
\end{equation}
where $F$ is the gravitational force acting between two objects, $m_1$ and $m_2$ are the masses of the objects, $r$ is the distance between the centers of their masses, and $G$ is the gravitational constant.
